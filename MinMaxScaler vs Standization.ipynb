```python
import pandas as pd
import numpy as np 
import random 
import matplotlib.pyplot as plt
```

```python
data =  list()
x = []
for i in range(19):
    m = random.randint(100, 1000)
    print(i ,  "==>", m)
    data.append(m)
```

```python
data = np.array(data)
data = data.reshape(-1, 1)
```

```python
## Normalization
from sklearn.preprocessing import MinMaxScaler
# define min max scaler
scaler = MinMaxScaler(feature_range=(-1, 1))
# transform data
normalized_arr = scaler.fit_transform(data)
```

```python
from sklearn.preprocessing import StandardScaler
object = StandardScaler()
scale = object.fit_transform(data) 
```

```python
df1 = pd.DataFrame(normalized_arr)
df2 = pd.DataFrame(scale)
df3 = pd.DataFrame(data)
df4 = pd.concat([df3, df1, df2], axis = 1)
df4 = df4.set_axis(['Original', 'MinMaxScaler', 'Standarization'], axis=1)
print(df4.describe())
print(df4)
```

```python
#plt.plot(df4.Original)
plt.plot(df4.MinMaxScaler, label = "xvdfd")
plt.plot(df4.Standarization)
# Display grid
plt.grid(True)
# Add title and labels
plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()  # display
```
